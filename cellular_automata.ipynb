```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
grid_N = 500

pop_p = 0.1

env = np.random.choice([1,0],size=(grid_N,grid_N),p=[pop_p,1-pop_p])
```

```python
plt.imshow(env);
```

```python
N_steps = 1000

ave_vec = np.array([0,0])

idx = np.random.choice(grid_N, size=2)

for s in range(N_steps):
    i,j = idx
    env[i%grid_N,j%grid_N] = 1
    
    idx += np.random.choice([-1,0,1],size=2)
    
    plt.imshow(env);
    plt.title(f'{s + 1} / {N_steps}')
    plt.show()
    
    clear_output(wait=True)
```
